```python
import matplotlib.pyplot as plt
import numpy as np
```

## 不同数量qubit下的BP问题

```python
loss_1 = np.load('data/original/loss/h1_4T30NDate100epoch300n1n_A3_0.npy')
loss_4 = np.load('data/original/loss/h1_4T30NDate100epoch300n4n_A3_0.npy')
loss_7 = np.load('data/original/loss/h1_4T30NDate100epoch300n7n_A3_0.npy')
loss_10 = np.load('data/original/loss/h1_4T30NDate100epoch300n10n_A3_0.npy')

x = np.arange(0, 300)

title = ['t=30', 't=29', 't=28', 't=27', 't=26']

fig, axs = plt.subplots(2, 3, figsize=(13, 8), sharex=True, sharey=True)
axs = axs.flatten() 

for i in range(0,5,1):
    axs[i].plot(x, loss_1[i], label='n=1')
    axs[i].plot(x, loss_4[i], label='n=4')
    axs[i].plot(x, loss_7[i], label='n=7')
    axs[i].plot(x, loss_10[i], label='n=10')
    axs[i].set_yscale('log')
    axs[i].annotate(title[i], xy=(0.5, -0.15), xycoords='axes fraction', ha='center', va='center')
    axs[i].legend(loc='lower left')

axs[5].axis('off')
    

plt.yscale('log')

fig.supxlabel(r'Iterations')
fig.supylabel(r'Loss')
plt.tight_layout()

plt.savefig('loss_iter_T_30_26.pdf', format='pdf', bbox_inches='tight')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

grad_norm_1 = np.load('data/original/grad/h1_4T30NDate100epoch300n1n_A3_0.npy')
grad_norm_4 = np.load('data/original/grad/h1_4T30NDate100epoch300n4n_A3_0.npy')
grad_norm_7 = np.load('data/original/grad/h1_4T30NDate100epoch300n7n_A3_0.npy')
grad_norm_10 = np.load('data/original/grad/h1_4T30NDate100epoch300n10n_A3_0.npy')

x = np.arange(0, 300)

title = ['t=30', 't=29', 't=28', 't=27', 't=26']

fig, axs = plt.subplots(2, 3, figsize=(13, 8), sharex=True, sharey=True)
axs = axs.flatten()  

for i in range(5):
    axs[i].plot(x, grad_norm_1[i*300:(i+1)*300], label='n=1')
    axs[i].plot(x, grad_norm_4[i*300:(i+1)*300], label='n=4')
    axs[i].plot(x, grad_norm_7[i*300:(i+1)*300], label='n=7')
    axs[i].plot(x, grad_norm_10[i*300:(i+1)*300], label='n=10')
    axs[i].set_yscale('log')
    axs[i].annotate(title[i], xy=(0.5, -0.15), xycoords='axes fraction', ha='center', va='center')
    axs[i].legend(loc='upper right')

axs[5].axis('off')

fig.supxlabel(r'Iterations')
fig.supylabel(r'Gradient norm')
plt.tight_layout()

plt.savefig('grad_iter_T_30_26.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
grad_norm_1 = grad_norm = np.load('data/original/grad/h1_4T30NDate100epoch300n1n_A3_0.npy')
grad_norm_4 = grad_norm = np.load('data/original/grad/h1_4T30NDate100epoch300n4n_A3_0.npy')
grad_norm_7 = grad_norm = np.load('data/original/grad/h1_4T30NDate100epoch300n7n_A3_0.npy')
grad_norm_10 = grad_norm = np.load('data/original/grad/h1_4T30NDate100epoch300n10n_A3_0.npy')

x = np.array([1, 4, 7, 10])

y=np.zeros(shape=(5, 4))

for i in range(5):
    y[i][0]=np.mean(grad_norm_1[i*300: (i+1)*300])
    y[i][1]=np.mean(grad_norm_4[i*300: (i+1)*300])
    y[i][2]=np.mean(grad_norm_7[i*300: (i+1)*300])
    y[i][3]=np.mean(grad_norm_10[i*300: (i+1)*300])

title = ['t=30', 't=29', 't=28', 't=27', 't=26']

fig, axs = plt.subplots(1, 5, figsize=(16, 4), sharex=True, sharey=True)

for i in range(5):
    axs[i].plot(x, y[i], lw=2)
    axs[i].set_yscale('log')
    axs[i].annotate(title[i], xy=(0.5, -0.15), xycoords='axes fraction', ha='center', va='center')
    

plt.yscale('log')

fig.supxlabel(r'qubits')
fig.supylabel(r'Gradient norm')
plt.tight_layout()

plt.savefig('grad_qubits.pdf', format='pdf', bbox_inches='tight')

plt.show()
```

```python
original = np.load('data/original/loss/h1_4T30NDate100epoch300n10n_A3.npy').reshape(-1)
improved = np.load('data/improved/loss/h0_0T30NDate100epoch300n10n_A3.npy').reshape(-1)

x = np.arange(1, 30 * 300 + 1)

fig, ax = plt.subplots()

ax.plot(x, original, label='original')
ax.plot(x, improved, label='improved')

fig.supxlabel(r'Iterations')
fig.supylabel(r'Loss')

ax.legend(loc='lower right')

plt.savefig('improved_original_loss.pdf', format='pdf', bbox_inches='tight')

plt.show()
```

```python
grad_norm_or = np.load('data/original/grad/h1_4T30NDate100epoch300n10n_A3.npy')
grad_norm_im = np.load('data/improved/grad/h0_0T30NDate100epoch300n10n_A3.npy')

x = np.arange(0, 300)

title = ['t=29', 't=29', 't=28', 't=27', 't=26']

fig, axs = plt.subplots(2, 3, figsize=(13, 8), sharex=True, sharey=True)
axs = axs.flatten() 

for i in range(0,5,1):
    axs[i].plot(x, grad_norm_or[i*300:(i+1)*300], label='original QuDDPM')
    axs[i].plot(x, grad_norm_im[i*300:(i+1)*300], label='improved QuDDPM')
    axs[i].set_yscale('log')
    axs[i].annotate(title[i], xy=(0.5, -0.15), xycoords='axes fraction', ha='center', va='center')
    axs[i].legend(loc='upper left')
    
axs[5].axis('off')

plt.yscale('log')

fig.supxlabel(r'Iterations')
fig.supylabel(r'Gradient norm')
plt.tight_layout()


plt.savefig('improved_original_grad.pdf', format='pdf', bbox_inches='tight')

plt.show()
```

```python
original = np.load('data/original/kl/h1_4T30NDate100epoch300n10n_A3.npy')
improved = np.load('data/improved/kl/h0_0T30NDate100epoch300n10n_A3.npy')
original = np.flip(original)
improved = np.flip(improved)

x = np.arange(0, 30)

fig, ax = plt.subplots()

ax.plot(x, original, label='original')
ax.plot(x, improved, label='improved')

fig.supxlabel(r'Iterations')
fig.supylabel(r'KL-Divergence')

plt.savefig('improved_original_grad.pdf', format='pdf', bbox_inches='tight')

ax.legend(loc='upper right')

fig.subplots_adjust(left=0.16, right=0.95, top=0.95, bottom=0.15, hspace=0.3, wspace=0.3)

plt.savefig('improved_original_kl.pdf', format='pdf', bbox_inches='tight')

plt.show()
```
